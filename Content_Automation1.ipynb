**Step 3:** Create Content Agents

```python
# --------------------------------------------------
# AGENT 1 — Topic Generator
# --------------------------------------------------
topic_agent = Agent(
    role="Topic Generator",
    goal="Generate 3 trending blog topics in the given niche.",
    backstory="Expert in identifying engaging blog topics.",
    llm=groq_llm,
    verbose=True,
    allow_delegation=False
)

# --------------------------------------------------
# AGENT 2 — Writer
# --------------------------------------------------
writer_agent = Agent(
    role="Content Writer",
    goal="Write an engaging and structured blog article.",
    backstory="Professional blog writer skilled in SEO content.",
    llm=groq_llm,
    verbose=True,
    allow_delegation=False
)

# --------------------------------------------------
# AGENT 3 — Editor
# --------------------------------------------------
editor_agent = Agent(
    role="Content Editor",
    goal="Improve clarity, grammar and structure.",
    backstory="Senior editor with expertise in refining content.",
    llm=groq_llm,
    verbose=True,
    allow_delegation=False
)
```

**Explanation:**

*  Topic Generator: Creates 3 trending blog topics for a niche.

*  Writer Agent: Writes structured, engaging blog content.
*  Editor Agent: Refines content for clarity, grammar, and flow.


**Step 4:** Create Topic Task

```python
# --------------------------------------------------
# USER INPUT
# --------------------------------------------------
niche = "AI Agents in Business Automation"

# --------------------------------------------------
# TASK 1 — Generate Topics
# --------------------------------------------------
topic_task = Task(
    description=f"""
Generate 3 trending blog topics in this niche:
{niche}

Return title + short explanation.
""",
    expected_output="Three blog topic ideas with explanation.",
    agent=topic_agent
)
```

**Explanation:**

*  Input: Provide niche (AI Agents in Business Automation).

*   Task: Generate 3 trending blog topics with short explanations.
*  Output: Structured list of topic ideas for content creation.


**Step 4:** Create Topic Task

```python
# --------------------------------------------------
# USER INPUT
# --------------------------------------------------
niche = "AI Agents in Business Automation"

# --------------------------------------------------
# TASK 1 — Generate Topics
# --------------------------------------------------
topic_task = Task(
    description=f"""
Generate 3 trending blog topics in this niche:
{niche}

Return title + short explanation.
""",
    expected_output="Three blog topic ideas with explanation.",
    agent=topic_agent
)
```

**Explanation:**

*  Niche Input: Focus on "AI Agents in Business Automation".

* Task Purpose: Generate 3 trending blog topics with short explanations.
*  Expected Output: Structured list of topic ideas for writing.



**Step 5:** Create Writing & Editing Tasks

```python
# --------------------------------------------------
# TASK 2 — Write Blog
# --------------------------------------------------
write_task = Task(
    description="""
Select the best topic from previous output.
Write a 600-800 word blog article including:
- Catchy Title
- Introduction
- Subheadings
- Conclusion
""",
    expected_output="Full structured blog article.",
    agent=writer_agent
)

# --------------------------------------------------
# TASK 3 — Edit Blog
# --------------------------------------------------
edit_task = Task(
    description="""
Improve grammar, clarity, and readability.
Return final polished version.
""",
    expected_output="Final edited blog article.",
    agent=editor_agent
)
```

**Explanation:**

*  Write Task: Select best topic and draft a 600–800 word blog with structure.

*   Edit Task: Refine grammar, clarity, and readability.
*  Output: Final polished blog article ready for publication.


**Step 6:** Launch Content Crew

```python
# --------------------------------------------------
# CREATE CREW
# --------------------------------------------------
crew = Crew(
    agents=[topic_agent, writer_agent, editor_agent],
    tasks=[topic_task, write_task, edit_task],
    verbose=True
)

# --------------------------------------------------
# RUN CREW
# --------------------------------------------------
result = crew.kickoff()

print("\n========== FINAL BLOG OUTPUT ==========\n")
print(result)
```

**Explanation:**

*   Assemble Crew: Combine topic generator, writer, and editor agents.

*   Execute Tasks: Run tasks sequentially—generate topics → write blog → edit content.
*   Output: Kickoff workflow and display the final polished blog article.
