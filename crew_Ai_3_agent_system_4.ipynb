**STEP 3:** Import & Create Custom API Tool

```python
from crewai import Agent, Task, Crew, Process
from crewai.tools import tool
import requests
import json


@tool("market_data_fetcher")
def fetch_market_data(dummy: str = "btc"):
    """
    Fetches Bitcoin and Ethereum prices from CoinGecko API.
    """
    url = "https://api.coingecko.com/api/v3/simple/price"

    params = {
        "ids": "bitcoin,ethereum",
        "vs_currencies": "usd"
    }

    response = requests.get(url, params=params)
    data = response.json()

    return json.dumps(data)
```

**EXPLANATION:**


*   Defines a custom CrewAI tool for fetching live crypto prices.

*  Connects to CoinGecko API to retrieve Bitcoin and Ethereum values in USD.
*   Converts API response into JSON format for agent processing.


**STEP 4:** Researcher Agent Creation

```python
researcher = Agent(
    role="Market Researcher",
    goal="Fetch real-time cryptocurrency data",
    backstory="Expert at collecting financial data from APIs.",
    tools=[fetch_market_data],
    llm="groq/llama-3.1-8b-instant",
    allow_delegation=False,
    verbose=True
)
```

**EXPLANATION:**


* Creates a market research agent for crypto data collection.

* Connects API tool to fetch real-time market prices.
* Uses Groq LLaMA model for intelligent processing.

*  Works independently with execution logs enabled.


**STEP 5:** Analyst Agent Creation

```python
analyst = Agent(
    role="Financial Analyst",
    goal="Analyze cryptocurrency prices and derive insights",
    backstory="Professional analyst who interprets market data.",
    llm="groq/llama-3.1-8b-instant",
    allow_delegation=False,
    verbose=True
)
```

**EXPLANATION:**

*  Creates a financial analyst agent for market interpretation.

*  Uses Groq LLaMA model to generate analytical insights.
*  Focused only on analysis (no tools attached).


* Runs independently with detailed logs enabled.

**STEP 6:** Reporter Agent Creation


```python
reporter = Agent(
    role="Financial Reporter",
    goal="Create a professional structured market report",
    backstory="Expert financial journalist who writes structured reports.",
    llm="groq/llama-3.1-8b-instant",
    allow_delegation=False,
    verbose=True
)
```

**EXPLANATION:**

*  Creates a reporting agent specialized in structured financial content.

*  Uses Groq LLaMA model to generate professional market reports.
*  Focused only on report writing (no external tools attached).

*  Works independently with execution logs enabled.


**STEP 7:** Research Task Creation

```python
research_task = Task(
    description="""
    Use the Market Data Fetcher tool to get the latest Bitcoin and Ethereum prices.
    Return the raw JSON data clearly.
    """,
    expected_output="JSON data of crypto prices.",
    agent=researcher
)
```

**EXPLANATION:**

*  Creates a task to fetch live crypto prices using the custom tool.

*  Instructs the agent to return raw JSON data without analysis.
*  Assigned to the researcher agent for real-time data collection.



**STEP 8:** Analysis Task Creation

```python
analysis_task = Task(
    description="""
    Analyze the cryptocurrency price data from the researcher.
    Compare Bitcoin and Ethereum.
    Identify which is priced higher.
    Provide 3 financial insights.
    """,
    expected_output="Detailed comparison and insights.",
    agent=analyst
)
```

**EXPLANATION:**

*  Creates a task to analyze fetched crypto price data.

*  Instructs comparison between Bitcoin and Ethereum.
* Requires identification of higher-priced asset and financial insights.


* Assigned to the analyst agent for interpretation and reasoning.

**STEP 9:** Report Task Creation

```python
report_task = Task(
    description="""
    Create a professional financial report using the analyst insights.
    Structure it with:
    - Title
    - Market Overview
    - Key Insights
    - Conclusion
    """,
    expected_output="Well-structured financial report.",
    agent=reporter
)
```

**EXPLANATION:**

*  Creates a task to convert analysis into a structured financial report.

*   Defines clear sections for professional formatting.
*   Assigned to the reporter agent for final content generation.




**STEP 10:** Full Workflow Execution

```python
crew = Crew(
    agents=[researcher, analyst, reporter],
    tasks=[research_task, analysis_task, report_task],
    process=Process.sequential,
    verbose=True
)

try:
    result = crew.kickoff()
    print(result)
except Exception as e:
    import traceback
    traceback.print_exc()

print("\nFINAL REPORT:\n")
```

**EXPLANATION:**

*   Creates a multi-agent crew with researcher, analyst, and reporter.

*   Runs tasks sequentially: data fetch в†’ analysis в†’ report generation.
*  Executes the complete workflow and prints the final structured financial report.
