**Step 4:** Create Agents

```python
# -----------------------------
# 1️ Research Agent
# -----------------------------
research_agent = Agent(
    role="Market Research Specialist",
    goal="Collect market data and industry insights.",
    backstory="Expert in gathering competitor data and trends.",
    llm=groq_model,
    verbose=True
)
# -----------------------------
# 2️ Trend Analyst
# -----------------------------
trend_analyst = Agent(
    role="Trend Analyst",
    goal="Analyze trends and forecast growth.",
    backstory="Specialist in interpreting market patterns.",
    llm=groq_model,
    verbose=True
)
# -----------------------------
# 3️ Report Generator
# -----------------------------
report_generator = Agent(
    role="Business Report Generator",
    goal="Create structured professional reports.",
    backstory="Writes executive-level summaries.",
    llm=groq_model,
    verbose=True
)
```

**Explanation:**

* Research Agent: Collects market data and insights.

*   Trend Analyst: Analyzes trends and forecasts growth.
*   Report Generator: Produces structured, executive-level reports.


**Step 5:** Create Tasks

```python
# -----------------------------
# TASKS
# -----------------------------
topic = "Electric Vehicle Market in India"
research_task = Task(
    description=f"Research market size, competitors, trends and policies of {topic}.",
    expected_output="Detailed research findings including market size, competitors, trends, and policies.",
agent=research_agent
)

analysis_task = Task(
    description="Analyze research and identify growth trends, risks and 3-year outlook.",
    expected_output="Structured analysis including growth trends, risks, opportunities and forecast.",
    agent=trend_analyst
)

report_task = Task(
    description="Generate a professional structured business report.",
    expected_output="Complete business report with executive summary, findings, trends, risks and conclusion.",
    agent=report_generator
)
```

**Explanation:**

*   Research Task: Gather detailed market data on the EV sector.


*   Analysis Task: Identify trends, risks, and 3-year forecasts.
*   Report Task: Produce a structured, executive-level business report.



**Step 6:** Launch Crew

```python
# -----------------------------
# CREW
# -----------------------------

crew = Crew(
    agents=[research_agent, trend_analyst, report_generator],
    tasks=[research_task, analysis_task, report_task],
    verbose=True
)

result = crew.kickoff()
```

**Explanation:**

*   Assemble Crew: Combine research, analysis, and report agents.
*   Execute Tasks: Run tasks in sequence to fetch, analyze, and report data.
*  Output: Kickoff workflow and print the final structured report.
