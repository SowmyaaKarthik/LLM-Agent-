**STEP 3:** Import & Tool Setup

```python
from crewai import Agent, Task, Crew, Process
from crewai_tools import SerperDevTool

search_tool = SerperDevTool()
```

**EXPLANATION:**
* CrewAI imports → Agent workflow components loaded

* Process module → Execution strategy support
*  Serper tool import → Web search capability added


* Search tool initialized → Tool ready for agent use



**STEP 4:** Research Agent Creation

```python
#  Research Agent (manual tool usage)
research_agent = Agent(
    role="AI Research Specialist",
    goal="Research detailed information about AI trends",
    backstory="Expert web researcher who gathers live AI information.",
    tools=[search_tool],
    llm="groq/llama-3.1-8b-instant",
    allow_delegation=False,
    verbose=True
)
```

**EXPLANATION:**


* Creates a specialized AI research agent focused on AI trends.

*  Connects Groq LLaMA model and enables web search capability.
*  Runs independently with detailed execution logs enabled.


**STEP 5:** Summarizer Agent Creation

```python
#  Summarizer Agent
summarizer_agent = Agent(
    role="Content Summarizer",
    goal="Summarize research clearly",
    backstory="Professional technical writer.",
    llm="groq/llama-3.1-8b-instant",
    allow_delegation=False,
    verbose=True
)
```

**EXPLANATION:**


*  Creates a summarizer agent to convert research into clear content.
*  Uses Groq LLaMA model for generating structured summaries.

* Works independently with verbose logging enabled.


**STEP 6:** Research Task Creation

```python
#  Research Task
research_task = Task(
    description="""
Search the internet for latest AI trends in 2026.
Use the search tool.
Provide detailed findings.
""",
    expected_output="Detailed AI trend report with examples.",
    agent=research_agent
)
```

**EXPLANATION:**

*  Creates a research task focused on latest AI trends.
*   Instructs the agent to use web search and provide detailed findings.

*  Assigns the task to the research agent for execution.


**STEP 7:** Summary Task Creation

```python
#Summary Task
summary_task = Task(
    description="""
Summarize the research into 5 clear bullet points with headings.
""",
    expected_output="Structured bullet summary.",
    agent=summarizer_agent
)
```

**EXPLANATION:**


*  Creates a summarization task based on research output.
*  Instructs to generate 5 clear bullet points with headings.


*  Assigns the task to the summarizer agent for structured output.
*   List item









**STEP 8:** Crew Execution Setup

```python
#  Crew Execution
crew = Crew(
    agents=[research_agent, summarizer_agent],
    tasks=[research_task, summary_task],
    process=Process.sequential,
    verbose=True
)
```

**EXPLANATION:**

*  Creates a crew with both research and summarizer agents.

*   Assigns research task first, then summary task.
*  Uses sequential process so tasks run step-by-step in order.




**STEP 9:** Final Execution

```python
result = crew.kickoff()
print("\n\nFINAL OUTPUT:\n")
print(result)
```

**EXPLANATION:**


*  Starts the complete multi-agent workflow.

* Executes research first, then summarization.
*   Prints the final structured output.
